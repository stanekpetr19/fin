```python
def wait_for_change(widget, value):
    future = asyncio.Future()
    def getvalue(change):
        # make the new value available
        future.set_result(change.new)
        widget.unobserve(getvalue, value)
    widget.observe(getvalue, value)
    return future
```

```python
#cleans input
def clean(df):
    '''accepts pandas.DataFrame (from Skandia!!!) as argument and cleans it + creates col for months'''
    
    df.columns = df.iloc[2]
    df = df.drop([0,1,2])
    df = df.reset_index()
    df = df.drop(columns = 'index')

    df.columns = ['date', 'desc', 'figure', 'balance']
    df.loc[df['desc'].str.contains('^2020|^2019', 
                                     flags = re.I, 
                                     regex = True, 
                                     na = False), 
            'desc'] = df['desc'].str.slice(start = 10)
    
    df['month'] = df['date'].str.split('-', expand = True)[1]
    cols = list(df.columns)
    df = df[[cols[0]] + [cols[-1]] + cols[1:-1]]

    
    return df
```
